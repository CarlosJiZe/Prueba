```python
def p1(sp1,sp2,sp3,sp4,numshares,numoptions,strike,tinterest,remaining,totinvest):
  print('Tabla 1 ---------------')
  all_stocks=[]
  all_stocks_02=[]
  all_options=[]
  all_options_02=[]
  bills=[]
  bills_02=[]
  price=[sp1,sp2,sp3,sp4]
  for i in price:
    i100=i*100
    all_stocks.append(i100)
  print(f'All stocks: {all_stocks}')
  for i in price:
    multi=(i-strike)*numoptions
    if multi < 0:
      all_options.append(0)
    else:
      all_options.append(multi)
  print(f'All options: {all_options}')
  for i in price:
    multi_02=(i-strike)*numshares
    if multi_02 < 0:
      resi=remaining*(1+(tinterest/100))
      bills.append(resi)
    else:
      resi=(remaining*(1+(tinterest/100)))+multi_02
      bills.append(resi)
  print(f'Bills + 100 options: {bills}')
  print('Tabla 2 ---------------')
  for i in all_stocks:
    resulti1=((i-totinvest)/totinvest)*100
    roundresi1=round(resulti1,0)
    all_stocks_02.append(roundresi1)
  print(f'All stocks 02: {all_stocks_02}')
  for i in all_options:
    resulti2=((i-totinvest)/totinvest)*100
    roundresi2=round(resulti2,0)
    all_options_02.append(roundresi2)
  print(f'All options 02: {all_options_02}')
  for i in bills:
    resulti3=((i-totinvest)/totinvest)*100
    roundresi3=round(resulti3,2)
    bills_02.append(roundresi3)
  print(f'Bills + 100 options 02: {bills_02}')

```

# **Pregunta_2**

The common stock of the P.U.T.T. Corporation has been trading in a narrow price range for the past month, and you are convinced it is going to break far out of that range in the next 3 months. You do not know whether it will go up or down, however. The current price of the stock is 130 per share, and the price of a 3-month call option at an exercise price of 130 is $10.85.

a. If the risk-free interest rate is 7% per year, what must be the price of a 3-month put option on P.U.T.T. stock at an exercise price of $130? (The stock pays no dividends.) (Do not round intermediate calculations. Round your answer to 2 decimal places.)

b. What would be a simple options strategy to exploit your conviction about the stock price’s future movements. How far would it have to move in either direction for you to make a profit on your initial investment? (Round your intermediate calculations and final answer to 2 decimal places.)

```python
def p2(costo,current_price_s0,stike_price,tasa_int,meses_expiración):
  T= meses_expiración/12
  porcentage=tasa_int/100
  s1=costo-current_price_s0
  s2=(stike_price)/((1+porcentage)**T)
  s3=s1+s2
  print(f'La respuesta al inciso a es: {round(s3,2)}')

  s4=s3+costo
  print(f'La respuesta al inciso b es: un straddle y {round(s4,2)}')
```

Imagine that you are holding 5,000 shares of stock, currently selling at 40 per share. You are ready to sell the shares but would prefer to put off the sale until next year for tax reasons. If you continue to hold the shares until January, however, you face the risk that the stock will drop in value before year-end. You decide to use a collar to limit downside risk without laying out a good deal of additional funds. January call options with a strike of 45 are selling at 2, and January puts with a strike price of 35 are selling at $3. Assume that you hedge the entire 5,000 shares of stock.

a. What will be the value of your portfolio in January (net of the proceeds from the options) if the stock price ends up at 30.

b. What will be the value of your portfolio in January (net of the proceeds from the options) if the stock price ends up at $40?

c. What will be the value of your portfolio in January (net of the proceeds from the options) if the stock price ends up at $50?

```python
def p12(numshares,s1,s2,s3,COST,POST,CS,PS):
  res=[]
  stock=[s1,s2,s3]
  initial_outlay=(numshares*CS)-(numshares*PS)
  for i in stock:
    sti=i*numshares
    if i < COST:
      resi=sti
    else:
      resi=sti-((COST-i)*5000)
    if i > POST:
      sumi=resi
    else:
      sumi=resi+((POST-i)*5000)
    resulti=sumi+initial_outlay
    print(f'El stock price a precio de {i} es de {resulti}')

    
```
